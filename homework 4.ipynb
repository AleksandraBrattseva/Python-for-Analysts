## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.0)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

```python
sns.jointplot(x=df['yr_built'], y=df['price'], kind='reg');
```

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['waterfront'], y=df['price'], whis=1.5)

plt.xlabel('waterfront')
plt.ylabel('price')
plt.title('Distribution of waterfront by price');
```

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['bathrooms'], y=df['price'], whis=1.5)

plt.xlabel('bathrooms')
plt.ylabel('price')
plt.title('Distribution of bathrooms by price');
```

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['view'], y=df['price'], whis=1.5)

plt.xlabel('view')
plt.ylabel('price')
plt.title('Distribution of view by price');
```
